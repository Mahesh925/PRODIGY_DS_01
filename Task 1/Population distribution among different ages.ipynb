```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib.ticker import FuncFormatter
```

```python
data = pd.read_csv("C:\Desktop\docs\Docs\Task 1\Dataset\population-by-age-group.csv")
```

```python
age_columns_pattern = ['Population by broad age group - Sex: all - Age: 65+ - Variant: estimates' , 'Population by broad age group - Sex: all - Age: 25-64 - Variant: estimates' , 'Population by broad age group - Sex: all - Age: 15-24 - Variant: estimates' , 'Population by broad age group - Sex: all - Age: 5-14 - Variant: estimates' , 'Population by broad age group - Sex: all - Age: 0-4 - Variant: estimates']
```

```python
selected_country = 'India'
```

```python
data_country = data[data['Entity'] == selected_country]
```

```python
population = data_country[age_columns_pattern].mean()
```

```python
custom_labels = ['65 and above' , '25-64' , '15-24' , '5-14' , '0-4']
```

```python
plt.figure(figsize = (12,7))
```

```python
plt.bar(custom_labels , population)
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: '{:,.0f}'.format(x)))
plt.title('Population distribution across age groups in India')
plt.xlabel('Age')
plt.ylabel('Population')
plt.show()
```
